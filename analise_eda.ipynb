```python
# Carregando a base de dados
df_precificacao = pd.read_csv('teste_indicium_precificacao.csv')
```

## 1.2: Limpeza dos dados

```python
# Verificando quantidade de valores faltantes em cada coluna
df_precificacao.isnull().sum()
```

```python
# Removendo colunar com mais de 30% de dados faltando
limite = len(df_precificacao) * 0.7
df_precificacao = df_precificacao.dropna(thresh=limite, axis=1)
```

```python
# Removendo duplicatas baseadas no id
df_precificacao = df_precificacao.drop_duplicates(subset=['id'])
```

## 1.3: Estabelecendo hip칩teses

### Hip칩tese 1: Os bairros localizados em Manhattan t칡m pre칞os mais elevados.

```python
# Calculando o pre칞o m칠dio por bairro_group (Top 10) e armazenando em um novo dataframe
df_pmbg = df_precificacao.groupby('bairro_group', as_index=False)['price'].median().sort_values('price', ascending=False).head(10)

# Criando um gr치fico em barras para visualiza칞칚o dos resultados acima
plt.figure(figsize=(12, 6))
sns.barplot(df_pmbg, x='bairro_group', y='price')
plt.xticks(rotation=45)
plt.title('Pre칞o M칠dio Por Bairro')
plt.show()
```

1. O gr치fico mostra que Manhattan possui o pre칞o m칠dio mais elevado entre os analisados. Isso confirma a hip칩tese que bairros localizados nessa regi칚o t칡m pre칞os mais caros.
2. H치 uma disparidade consider치vel de valores entre Manhattan e outras regi칫es, indicando uma valoriza칞칚o das propriedades nesse local.
3. Alguns fatores contribuintes para a confirma칞칚o da hip칩tese s칚o: oferta de terreno e imoveis mais limitada (por ser uma ilha), centro econ칪mico global, turismo e cultura, etc.

### Hip칩tese 2: Espa칞os do tipo casa inteira ou apartamento (Entire home/apt) s칚o mais caros.

```python
# Criando um gr치fico em boxplot por tipo de espa칞o
sns.boxplot(df_precificacao, x='room_type', y='price')
plt.ylim(0, 500) # Remove outliers
plt.title('Pre칞os Por Tipo de Espa칞o')
plt.show()
```

1. Os dados confirmam a hip칩tese de que espa칞os do tipo casa inteira ou apartamento (Entire home/apt) s칚o mais caros.
2. Espa칞os do tipo Entire home/apt t칡m mediana de pre칞o mais alta em compara칞칚o aos outros espa칞os.
3. Varia칞칚o maior de pre칞os em Entire home/apt. Provavelmente por diferen칞as de tamanho, localiza칞칚o e comodidades oferecidas.
4. Alguns outliers nas 3 cateorias. Maior quantidade em Entire home/apt, sugerindo que alguns espa칞os deste tipo podem ser excepcionalmente caros.

### Hip칩tese 3: Listagens com maior n칰mero de reviews t칡m pre칞os mais elevados.

```python
# Criando um gr치fico de dispers칚o com correla칞칚o entre o n칰mero de reviews e pre칞o
sns.scatterplot(df_precificacao, x='numero_de_reviews', y='price', alpha=0.5)
plt.title('Rela칞칚o Entre N칰mero de Reviews e Pre칞o')
plt.show()
```

1. As informa칞칫es n칚o confirmam a hip칩tese em quest칚o.

## 2.1: Supondo que uma pessoa esteja pensando em investir em um apartamento para alugar na plataforma, onde seria mais indicada a compra?

```python
# Calculando o pre칞o m칠dio e demanda (baseada em reviews) por bairro
df_analise_bairro = df_precificacao.groupby('bairro').agg(preco_medio=('price', 'median'), demanda=('numero_de_reviews', 'sum')).sort_values('demanda', ascending=False).head(10)

df_analise_bairro
```

## 2.2: O n칰mero m칤nimo de noites e a disponibilidade ao longo do ano interferem no pre칞o?

```python
# Verificando a correla칞칚o n칰mero m칤nimo de noites e disponibiliade no ano
df_corr_noites_disponibilidade = df_precificacao[['price', 'minimo_noites', 'disponibilidade_365']].corr()
sns.heatmap(df_corr_noites_disponibilidade, annot=True, cmap='coolwarm')
plt.title('Correla칞칫es Entre N칰mero de Noites e Disponibilidade')
plt.show()
```

## 2.3: Existe algum padr칚o no texto do nome do local para lugares de mais alto valor?

```python
# Filtrando os 10% locais mais caros
df_top10_caros = df_precificacao[df_precificacao['price'] >= df_precificacao['price'].quantile(0.9)]

# Contando as palavras no nome dos locais filtrados
contagem_palavras = df_top10_caros['nome'].str.split(expand=True).stack().value_counts()

# Coletando as 20 palavras mais comuns
palavras = contagem_palavras.head(20)

# Demonstrando em um gr치fico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=palavras.values, y=palavras.index, orient='h')
plt.xlabel('Frequ칡ncia')
plt.ylabel('Palavras')
plt.title('Palavras Mais Comuns (Locais de Alto Valor)')
plt.show()
```
